```python
import pandas as pd
import numpy as np
import re
```

```python
praisemageddon = pd.read_excel("data/Impact Hour Distribution analysis.xlsx", sheet_name="Praisemageddon", engine='openpyxl', header=0,  usecols="A,E").dropna()
```

### Resolving Names

```python
import pickle

with open('outputs/names_dict.pickle', 'rb') as f:
    names_dict = pickle.load(f)

praisemageddon.loc[:,"Name"] = praisemageddon.loc[:,"Name"].map(names_dict)

praisemageddon
```

```python
default = pd.read_csv('outputs/YGG-AUGMENTATION.csv')

default['Paid Ratio'] = default['Total Hours Deducted'] / -default['Default Net Impact Hours']

default['Unpaid Bonus Factor'] = 1 - (default['Paid Ratio'] / default['Paid Ratio'].sum() * 10)

default
```

```python
combined = default.merge(praisemageddon, left_on='Hatcher', right_on='Name')

combined
```

```python
combined['praisemaggedon'] = combined['Unpaid Bonus Factor'] * combined['IH held']

combined['praisemaggedon'] = combined['praisemaggedon'] / combined['praisemaggedon'].sum()

combined
```

```python
combined = combined.sort_values('praisemaggedon', ascending=False)[['Name', 'Initial Distribution', 'Unpaid Bonus Factor', 'praisemaggedon']].ffill()

combined['Final Distribution'] = combined['Initial Distribution'] * 0.52 + combined['praisemaggedon'] * 0.48

combined.sort_values('Final Distribution', ascending=False).head(40)
```

```python
combined.sort_values('Final Distribution', ascending=False).to_csv('outputs/COMBINED-WINNING-PROPOSALS.csv')
```
